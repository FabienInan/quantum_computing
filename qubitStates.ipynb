```python
from qiskit import QuantumCircuit, assemble, Aer
from qiskit.visualization import plot_histogram, plot_bloch_vector
from math import sqrt, pi
```

```python
qc = QuantumCircuit(1)
#initial_state = [0,1]   # Define initial_state as |1>
initial_state = [1/sqrt(2), 1j/sqrt(2)] # superposition state , equal probability
#initial_state = [1/sqrt(3),1/sqrt(3/2)] # superposition state , first qubit probability of 1/3
qc.initialize(initial_state, 0) # Apply initialisation operation to the 0th qubit
qc.draw()  # Let's view our circuit
```

```python
qc.save_statevector()   # Tell simulator to save statevector
sim = Aer.get_backend('aer_simulator')
result = sim.run(qc).result() # Do the simulation and return the result
out_state = result.get_statevector()
print(out_state) 
```

```python
qc.measure_all()
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc).result()
counts = result.get_counts()
plot_histogram(counts)
```
